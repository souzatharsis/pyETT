# Matchup and Cohort Analysis

```python
from pyETT import ett
```

The **ett** Class **Cohort** manages state, logic and analytics for a cohort of players. Let's take a look at how the top ETT players in the world match up against each other.
First, we create a cohort with the top 10 players from the leaderboard:

```python
lb_cohort = ett.Cohort(ett.ETT().get_leaderboard(num_players=10))
lb_cohort.size
```

```python
lb_cohort.players_dataframe()
```

By accessing the attribute ``size`` and calling the methods ``players_dataframe()`` from the **Cohort** object, we obtain the cohort size (10) and the list of players in the cohort.
We can get the Elo history of the players in the cohort with the method ``get_elo_history()``:

```python
lb_cohort_elo = lb_cohort.get_elo_history()
lb_cohort_elo 
```

```python
lb_cohort_elo.plot(figsize=(10,8))
```

We can obtain the full list of matches played within the cohort, i.e., match ups, by accessing the method ``get_matches()`` or ``get_matches_dataframe()`` to obtain results in a pandas DataFrame:

```python
lb_cohort.get_matches_dataframe()
```

We observed the the top 10 players played an impressive 443 matches among each other. To obtain match up analytics of players in the cohort, we can call the method ``describe()``:

```python
lb_cohort.describe()
```

We observe that the player ``lapoctopus90`` (ranked 2) played the most ranked matches against Top 10 ranked players with a total of 123 matches followed by ``twISTer1979`` with 110 matches. The player with the highest win rate when playing against other Top 10 ranked players is ``PROTTVR`` followed by number World's number 1 ranked player ``yedat``. Meanwhile, the player ``alexttbcn`` managed to reach the top of the leaderboard without playing many of the Top 10 ranked players with just 19 ranked matches against them.
